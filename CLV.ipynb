## BASIC CLV

```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv("merged_data.csv")

df.head()
```

```python
# 1️⃣ Toplam Harcama (Revenue) ve Sipariş Sayısı
customer_metrics = df.groupby("CustomerID").agg({
    "InvoiceNo": "nunique",   # Müşteri başına sipariş sayısı
    "Sales": "sum"            # Müşteri başına toplam harcama
}).rename(columns={"InvoiceNo": "customer_orders", "Sales": "total_spent"})
```

```python
print(customer_metrics)
```

```python
# 2️⃣ Ortalama Sipariş Değeri (AOV)
customer_metrics["AOV"] = customer_metrics["total_spent"] / customer_metrics["customer_orders"]

```

```python
print(customer_metrics)
```

```python
# 3️⃣ Satın Alma Sıklığı (Purchase Frequency - PF)
total_orders = df["InvoiceNo"].nunique()  # Toplam sipariş sayısı
total_customers = df["CustomerID"].nunique()  # Toplam müşteri sayısı
purchase_frequency = total_orders / total_customers  # Genel satın alma sıklığı

```

```python
print("Toplam Sipariş: ",total_orders)
print("Toplam Müşteri: ",total_customers)
print("Toplam Satın Alma Sıklığı: ", purchase_frequency)
```

```python
#1. Adım: Müşteri başına sipariş sayısını bulma
customer_orders = df.groupby('CustomerID')['InvoiceNo'].nunique()

# 2. Adım: Repeat Rate hesaplamak
# Repeat Rate, 1'den fazla sipariş veren müşterilerin oranı
repeat_customers = customer_orders[customer_orders > 1]
repeat_rate = len(repeat_customers) / len(customer_orders)

# 3. Adım: Müşteri Ömrü (LT) hesaplama
lt = 1 / (1 - repeat_rate)

# Sonuçları yazdırma
print(f"Repeat Rate: {repeat_rate:.2f}")
print(f"Müşteri Ömrü (LT): {lt:.2f} yıl")
```

```python
# 4️⃣ Müşteri Ömrü (Lifetime - LT) ➜ Ortalama 1 yıl sürede alışveriş yapan müşteri sayısına göre belirlenir
customer_metrics["LT"] = lt  # Basit varsayım: 1 yıl sürede müşteri kaybı yok
```

```python
# 5️⃣ Müşteri Ömür Değeri (CLV) Hesaplama
customer_metrics["CLV"] = customer_metrics["AOV"] * purchase_frequency * customer_metrics["LT"]
```

```python
# Sonuçları Göster
print(customer_metrics[["AOV", "customer_orders", "CLV"]].head())
```

```python
customer_metrics.reset_index(inplace=True)
```

# Percentile yöntemini kullanarak CLV'yi sınıflandırma

-> 80. yüzdelik değerin üstündeki müşteriler, tüm müşterilerin en yüksek CLV’ye sahip ilk %20’lik dilimi demektir. 20. yüzdelik değerin altındaki müşteriler ise en düşük CLV’ye sahip %20’lik dilim

```python
# CLV'nin %80'ini ve %20'sini bulma
percentile_80 = customer_metrics["CLV"].quantile(0.80)
percentile_20 = customer_metrics["CLV"].quantile(0.20)

# Yüksek ve Düşük CLV'yi sınıflandırma
customer_metrics['CLV_Category'] = customer_metrics['CLV'].apply(
    lambda x: 'High CLV' if x >= percentile_80 
    else ('Low CLV' if x <= percentile_20 
    else 'Medium CLV')
)

print(customer_metrics)
```

## Gamma-Gamma Modeli

```python
rfm_data = pd.read_csv("rfm_clv_data.csv")
```

```python
rfm_data.head()
```

```python
# NaN değerlerini kontrol et
nan_values = rfm_data.isna().sum()
print("Eksik (NaN) Değerler:\n", nan_values)
```

```python
zero_values = (rfm_data == 0).sum()
print("Sıfır Değerler:\n", zero_values)
```

```python
# UnitPrice, Sales, Monetary, total_spent, AOV ve CLV sütunlarında sıfır olan satırları kaldır
df_clean = rfm_data[(df['UnitPrice'] != 0) & (rfm_data['Sales'] != 0) & 
              (rfm_data['Monetary'] != 0) & (rfm_data['total_spent'] != 0) &
              (rfm_data['AOV'] != 0) & (rfm_data['CLV'] != 0)]

```

```python
zero_values = (df_clean == 0).sum()
print("Sıfır Değerler:\n", zero_values)
```

```python
rfm_clv_data = df_clean
```

```python
rfm_clv_data.head()
```

```python
print(rfm_clv_data["AOV"].dtype)
```

```python
rfm_clv_data['AOV'] = rfm_clv_data['AOV'].astype(int)
```

```python
print(rfm_clv_data['AOV'].dtype)
```

```python
rfm_clv_data
```

## CLV Hesaplaması

```python
# Her müşteri için toplam 'customer_orders' ve ortalama 'AOV' hesaplayalım
rfm_clv_data_grouped = rfm_clv_data.groupby('CustomerID').agg({
    'AOV': 'mean',  # Ortalamayı alıyoruz
    'customer_orders': 'sum'  # Toplam sipariş sayısını alıyoruz
}).reset_index()

# Yeni veri setini kontrol edelim
print(rfm_clv_data_grouped.head())
```

```python
from lifetimes import GammaGammaFitter

# Gamma-Gamma modelini başlatıyoruz:
ggf = GammaGammaFitter(penalizer_coef=0.001)

# Modeli fit ediyoruz:
ggf.fit(rfm_clv_data_grouped['AOV'], rfm_clv_data_grouped['customer_orders'])

# Her müşteri için beklenen ortalama işlem değerini hesaplayalım
rfm_clv_data_grouped["expected_monetary"] = ggf.conditional_expected_average_profit(
    rfm_clv_data_grouped['AOV'], rfm_clv_data_grouped['customer_orders']
)

# Sonuçları gösterelim
print(rfm_clv_data_grouped[['CustomerID', 'AOV', 'customer_orders', 'expected_monetary']].head())

```

```python

# AOV ve customer_orders için temel istatistikleri inceleyelim
print(rfm_clv_data_grouped[['AOV', 'customer_orders']].describe())
```

## Otulier Belirlenmesi ve Sadece Outlier Verilerle CLV

```python
from scipy.stats import zscore

# Z-score hesaplayalım
rfm_clv_data_grouped['AOV_zscore'] = zscore(rfm_clv_data_grouped['AOV'])
rfm_clv_data_grouped['customer_orders_zscore'] = zscore(rfm_clv_data_grouped['customer_orders'])

# Z-score'u 3'ten büyük olanları (outlier) filtreleyelim
outliers_zscore = rfm_clv_data_grouped[
    (rfm_clv_data_grouped['AOV_zscore'].abs() > 3) | 
    (rfm_clv_data_grouped['customer_orders_zscore'].abs() > 3)
]

# Outlier olan müşterileri listeleyelim
print(outliers_zscore[['CustomerID', 'AOV', 'customer_orders', 'AOV_zscore', 'customer_orders_zscore']])
```

```python
# Aykırı değerleri dışlayan yeni veri seti
rfm_clv_data_filtered = rfm_clv_data_grouped[
    (rfm_clv_data_grouped['AOV_zscore'].abs() <= 3) &
    (rfm_clv_data_grouped['customer_orders_zscore'].abs() <= 3)
]

# Yeni veriyi kontrol edelim
rfm_clv_data_filtered
```

```python
# Yeni veriyi kullanarak Gamma-Gamma modelini başlatıyoruz
ggf_filtered = GammaGammaFitter(penalizer_coef=0.001)

# Modeli fit ediyoruz
ggf_filtered.fit(rfm_clv_data_filtered['AOV'], rfm_clv_data_filtered['customer_orders'])

# Her müşteri için beklenen ortalama işlem değerini (expected_monetary) hesaplayalım
rfm_clv_data_filtered['expected_monetary'] = ggf_filtered.conditional_expected_average_profit(
    rfm_clv_data_filtered['AOV'], rfm_clv_data_filtered['customer_orders']
)

# Sonuçları görelim
print(rfm_clv_data_filtered[['CustomerID', 'AOV', 'customer_orders', 'expected_monetary']].head())
```

```python
print(rfm_clv_data_filtered[['AOV', 'customer_orders']].describe())
```

## Outlier Çıkartılmış Veri ile CLV 

```python
# Outlier olan müşteri ID'lerini alalım
outlier_customer_ids = outliers_zscore['CustomerID'].tolist()

# Listeyi görüntüleyelim
print(outlier_customer_ids)
```

```python
# Bu müşteri ID'lerine sahip satırları çıkaralım
rfm_clv_data_no_outliers = rfm_clv_data[~rfm_clv_data['CustomerID'].isin(outlier_customer_ids)]

# Sonuçları kontrol edelim
print(rfm_clv_data_no_outliers.head())
```

```python
# Her müşteri için toplam 'customer_orders' ve ortalama 'AOV' hesaplayalım
rfm_clv_data_no_outliers_grouped = rfm_clv_data_no_outliers.groupby('CustomerID').agg({
    'AOV': 'mean',  # Ortalamayı alıyoruz
    'customer_orders': 'sum'  # Toplam sipariş sayısını alıyoruz
}).reset_index()

# Yeni veri setini kontrol edelim
print(rfm_clv_data_no_outliers_grouped.head())
```

```python
from lifetimes import GammaGammaFitter

# Gamma-Gamma modelini başlatıyoruz:
ggf = GammaGammaFitter(penalizer_coef=0.001)

# Modeli fit ediyoruz:
ggf.fit(rfm_clv_data_no_outliers_grouped['AOV'], rfm_clv_data_no_outliers_grouped['customer_orders'])

# Her müşteri için beklenen ortalama işlem değerini hesaplayalım
rfm_clv_data_no_outliers_grouped["expected_monetary"] = ggf.conditional_expected_average_profit(
    rfm_clv_data_no_outliers_grouped['AOV'], rfm_clv_data_no_outliers_grouped['customer_orders']
)

# Sonuçları gösterelim
print(rfm_clv_data_no_outliers_grouped[['CustomerID', 'AOV', 'customer_orders', 'expected_monetary']].head())

```

```python

# AOV ve customer_orders için temel istatistikleri inceleyelim
print(rfm_clv_data_no_outliers_grouped[['AOV', 'customer_orders']].describe())
```
